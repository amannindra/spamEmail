```python
from annoy import AnnoyIndex
import random

f = 40
t = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for i in range(1000):
    v = [random.gauss(0, 1) for z in range(f)]
    t.add_item(i, v)

# print(len(v))
# print(v)


t.build(10, n_jobs=-1) # 10 trees, builds a forest of n_trees trees. More trees give a higher precision when queried
t.save('test.ann')

# # ...

u = AnnoyIndex(f, 'angular')
u.load('test.ann') # super fast, will just mmap the file
a = u.get_nns_by_item(0, 1000)
print(a)# will find the 1000 nearest neighbors
print(len(a))
```

```python
t = AnnoyIndex(40, 'angular') #Intializes the index with 40 dimensions

f = 40
for i in range(1000):
    v = [random.randint(0, 1000) for _ in range(f)]
    t.add_item(i, v)

t.build(10) # 10 trees,  The number of trees in the index. More trees mean higher accuracy but slower query time.\

distance_threshold = 0.5
s = 0
duplicates = []
for i in range(1000):
    n, d = t.get_nns_by_item(i, 1000, include_distances=True)
    for neighbor, distance in zip(n, d):
        if i != neighbor and distance < distance_threshold:
            duplicates.append((i, neighbor))
            s += 1

print(s)

# t.save('my_index.ann')  

# u = AnnoyIndex(40, "angular")
# u.load("my_index.ann")

# # Using get_nns_by_item
# print(t.get_nns_by_item(0,10))

```

```python
from annoy import AnnoyIndex

# 3 dimensional vector
a = AnnoyIndex(3, 'angular')


#add_item(item, vector)
#Add a single item to the index. The item is a zero-based index that will be used to identify the vector. The vector is a list of floats.


a.add_item(0, [1, 0, 0])
a.add_item(1, [0, 1, 0])
a.add_item(2, [0, 0, 1])


#build(n_trees, n_jobs) n_job = -1 means use all available cores
#Builds a forest of n_trees trees. More trees gives higher precision when querying. After calling this method, no more items may be added.
a.build(-1)
a.save('test.tree')


# 3 dimensional vector
b = AnnoyIndex(3)
b.load('test.tree')


# get_nns_by_item(item, n, search_k=-1, include_distances=False)
# Returns the indices of the n nearest neighbors to item.

print(b.get_nns_by_item(0, 100))
# print(b.get_nns_by_vector([1.0, 0.5, 0.5], 100))
```
